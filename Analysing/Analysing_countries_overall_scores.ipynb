```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Main/main.csv')
```

```python
df.head()
```

Ordering by New business density rate.

```python
df_nbd = df.groupby('Country').agg({'New business density rate': 'mean'})
```

```python
df_nbd.sort_values(by='New business density rate', ascending=False).head(50)
```

Ordering by Economic Freedom index overall score.

```python
df_os = df.groupby('Country').agg({'Economic freedom index': 'mean'})
```

```python
df_os.sort_values(by='Economic freedom index', ascending=False).head(50)
```
