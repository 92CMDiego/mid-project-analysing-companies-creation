```python
import pandas as pd
import numpy as np
import pycountry
```

```python
df_efi = pd.read_csv('raw_economic_freedom_scores_1995_2023.csv')
```

```python
df_efi.head()
```

```python
df_efi = df_efi.drop('Id', axis=1)
```

```python
df_efi.rename(columns={'ISO Code': 'ISO_code_alpha2'}, inplace=True)
```

```python
iso3_mapping = {country.alpha_2: country.alpha_3 for country in pycountry.countries}
```

```python
df_efi['ISO_code_alpha3'] = df_efi['ISO_code_alpha2'].map(iso3_mapping)
```

```python
df_efi
```

```python
df_efi.isna().sum()
```

```python
len(df_efi)
```

```python
df_efi[df_efi['ISO_code_alpha3'].isna()==True]['Short Name'].unique()
```

```python
df_efi[df_efi['Overall Score'].isna()==True]['Short Name'].unique()
```

```python
df_efi = df_efi.dropna(subset=['ISO_code_alpha3'])
```

```python
df_efi = df_efi.dropna(subset=['Overall Score'])
```

```python
df_efi.isna().sum()
```

```python
df_efi = df_efi.drop('Short Name', axis=1)
```

```python
df_efi.rename(columns={'Name': 'Country'}, inplace=True)

col_order = ['Country', 'ISO_code_alpha3', 'ISO_code_alpha2', 'Index Year', 'Overall Score',
                'Property Rights', 'Judicial Effectiveness', 'Government Integrity',
                'Tax Burden', 'Government Spending', 'Fiscal Health',
                'Business Freedom', 'Labor Freedom', 'Monetary Freedom',
                'Trade Freedom', 'Investment Freedom', 'Financial Freedom']
df_efi = df_efi[col_order]
```

```python
df_efi['Rule of Law'] = np.where(df_efi['Judicial Effectiveness'].isna(), np.nan, (df_efi['Property Rights'] + df_efi['Government Integrity'] + df_efi['Judicial Effectiveness']) / 3)
df_efi['Government Size'] = np.where(df_efi['Fiscal Health'].isna(), np.nan, (df_efi['Government Spending'] + df_efi['Tax Burden'] + df_efi['Fiscal Health']) / 3)
df_efi['Regulatory Efficiency'] = (df_efi['Business Freedom'] + df_efi['Labor Freedom'] + df_efi['Monetary Freedom']) / 3
df_efi['Open Markets'] = (df_efi['Trade Freedom'] + df_efi['Investment Freedom'] + df_efi['Financial Freedom']) / 3
```

```python
df_efi['Rule of Law'] = df_efi['Rule of Law'].round(2)
df_efi['Government Size'] = df_efi['Government Size'].round(2)
df_efi['Regulatory Efficiency'] = df_efi['Regulatory Efficiency'].round(2)
df_efi['Open Markets'] = df_efi['Open Markets'].round(2)
```

```python
df_efi
```

```python
df_efi[df_efi.select_dtypes(include=['float']).columns] = df_efi.select_dtypes(include=['float']).round(2)
```

```python
df_efi.dtypes
```

```python
df_efi.insert(0, 'id', df_efi['ISO_code_alpha3'] + '-' + df_efi['Index Year'].astype(str))
```
